## Predict the percentage of students based on No. of Hours studied

### Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
df.head()
```

```python
#performing some EDA(Exploratory data analysis) on dataset
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.sample(2)
```

```python
df.dtypes
```

```python
df.Hours
```

```python
#splitting the dataset into X and y
X = df[['Hours']]
y = df['Scores']
```

```python
#applying train_test_split on data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
#Building ML model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
#training the model with the necessary data
model.fit(X_train,y_train)
```

```python
model.coef_  #coeffecient of model
```

```python
model.intercept_  # intercept of model
```

```python
#prediction of model
y_pred = model.predict(X_test)
y_pred
```

```python
#evaluate the model
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,accuracy_score
print("mean squared error : ",mean_squared_error(y_test,y_pred))
print("mean absolute error: ",mean_absolute_error(y_test,y_pred))
print("root mean  squared error: ",np.sqrt(mean_squared_error(y_test,y_pred)))
print("r2_score : ",r2_score(y_test,y_pred)*100)
```
